#### Display the trending Python repositories in GitHub

The steps to solve this problem is similar to the previous one except that you need to find out the repository names instead of developer names.

````python
# This is the url you will scrape in this exercise
url = 'https://github.com/trending/python?since=daily'
html=requests.get(url).content
soup=BeautifulSoup(html,"lxml")
soup
````

#### Retrieve an arbitary Wikipedia page of "Python" and create a list of links on that page

````python
# This is the url you will scrape in this exercise
url ='https://en.wikipedia.org/wiki/Python' 
html=requests.get(url).content
soup=BeautifulSoup(html,"lxml")
soup
````

````python
#your code
text = [element for element in soup.select("li>a")]
text1=[texts["href"] for texts in text]
text1 
links=["https://en.wikipedia.org" + em if em.startswith("/") else em for em in text1]
links1=["https://en.wikipedia.org/wiki/Python" + em if em.startswith("#") else em  for em in links]
links1
````

## Display the date, days, title, city, country of next 25 hackathon events as a Pandas dataframe table

````python
# This is the url you will scrape in this exercise
url ='https://mlh.io/seasons/na-2020/events'
html=requests.get(url).content
soup=BeautifulSoup(html,"lxml")
soup
````

````python
#your code
date=[element.text.strip() for element in soup.select("p.event-date")]
title= [element.text.strip() for element in soup.select("h3.event-name")]
local=[element.text.strip().replace("\n         ", "").split(",") for element in soup.select("div.event-location")]
city=[local[i][0] for i in range(len(local)) ]
country=[local[i][1] for i in range(len(local)) ]
df=pd.DataFrame()
df["date"]=date
df["title"]=title
df["city"]=city
df["country"]=country
df.head(25)
````
